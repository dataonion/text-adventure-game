```
import time  # import time to use sleep function
import random


def print_pause(message_to_print):
    print(message_to_print)
    time.sleep(2)


def intro():
    print_pause("Gasp!")
    print_pause("You open your eyes and realized that "
                "there is fire all around you.")
    print_pause("I remember now...I was just fighting my arch enemy "
                "Drakan and hit on the head and collapsed.")
    print_pause("You realize that the house you are in is in "
                "flames and need to get you as soon as possible.")
    print_pause("You see that your room door is blocked with burning wood.")


def valid_input(prompt, option1, option2):
    while True:
        response = input(prompt).lower()
        if response == option1:
            break
        elif response == option2:
            break
        else:
            print_pause("That is not an option.")
    return response


def second_challenge():
    print_pause("You managed to open the door.")
    print_pause("but as soon as you got out, you see a demon "
                "waiting to fight you, probably Drakan's underlings.")
    print_pause("You draw your staff, and start casting a spell.")
    print_pause("You start casting a spell.")
    print_pause("You recently learned this spell, so you might"
                "need to try a few times to get it working,"
                "but you do it anyway.")


def end():
    print_pause("")
    print_pause("As you step outside the burning house,"
                "your eyes were struck with bright sunlight as if"
                "celebrating your successful escape.")
    print_pause("Drakan took away your magic talisman "
                "and you will need to get it back.")
    print_pause("but for now, that is yet another tale "
                "we will need to get into later on.")
    print_pause("The house started collapsing slowly, and "
                "with a deep sigh, you start walking towards the sun, ")
    print_pause("hoping to find your next way and to fight Drakan again...")
    print_pause("===To be Continued===")


while True:
    while True:
        intro()

        print_pause("What will you do?")

        print("1: Kick the wood blocking the door.")
        print("2: Use your water spell to put out the fire on the door.")

        response = valid_input("", '1', '2')

        if response == '1':
            print_pause("you broke your foot.")
        elif response == '2':
            print_pause("you put out the fire with water.")

        second_challenge()

        print_pause("The staff starts glowing")
        demon_hp = 100

        while demon_hp >= 0:
            dmg = random.randint(5, 25)
            print("demon health is " + str(demon_hp))
            demon_hp -= dmg
            print("You continue attacking, giving " + str(dmg) + " damage!")

        print_pause("As you dealt your last blow of frostbolt "
                    "at the demon, he slowly collapsed.")

        end()
        break
    while True:
        play_again = input("play again? yes or no  ").lower()
        if play_again == 'yes':
            print_pause("Very well. Let me tell you her story again.")
            print("")
            break
        elif play_again == 'no':
            print("Ok. Bye.")
            break
        else:
            print("Sorry I didn't get you.")
    if 'no' in play_again:
        break
```
